```python
"""1.1.10. What are global, protected and private attributes in Python?"""

class Variables:
    def __init__(self, a, b, c):
        self.a = a   # public data member
        self._b = b  # protected data member
        self.__c = c # private data member

    def accessPublicMembers(self):
        print("Public Data Member: ", self.a)

    def _accessProtectedMembers(self):
        print("Protected Data Member: ", self._b)
        
    def __accessPrivateMembers(self):
        print("Private Data Member: ", self.__c)

    def accessPrivateMembersInPublicFunction(self):
        self.__accessPrivateMembers()

class Sub(Variables):   #subclass to access protected member
    def __init__(self, a, b, c):
        Variables.__init__(self, a, b, c)
        
    def accessProtectedMembersInSubClass(self):
        self._accessProtectedMembers()

obj = Sub("hello", "there", "Neelanjana")
obj.accessPublicMembers()
obj.accessProtectedMembersInSubClass()
obj.accessPrivateMembersInPublicFunction()

# Object can access protected member
print("Object is accessing protected member:", obj._b)

# object can not access private member, so it will generate Attribute error
try:
    print(obj.__c)
except Exception as e:
    print(e)
```

```python
"""1.7. Write a program to create a simple calculator performing only four basic operations(+,-,/,*) ."""

import functools as f
class calculator:
    def addition(self,*arg):
        return("a+b=",f.reduce(lambda a,b: a+b,arg))
    def subtraction(self,*arg):
        return("a-b=",f.reduce(lambda a,b: a-b,arg))
    def multiplication(self,*arg):
        return("a*b=",f.reduce(lambda a,b: a*b,arg))
    def division(self,*arg):
        return("a/b=",f.reduce(lambda a,b: a/b,arg))
        
calc=calculator()
print(calc.addition(1,2,3,4,5))
print(calc.subtraction(1,2,3,4,5))
print(calc.multiplication(1,2,3,4,5))
print(calc.division(1,2))
```

```python
"""1.10. Write a program to find the sum of all the positive numbers entered by the user. As soon as the user
enters a negative number, stop taking in any further input from the user and display the sum ."""

import functools as f
def add(*arg):
    pos_num=[]
    for num in arg:
        if num>0:
            pos_num.append(num)
        elif num<0:
            break
    return(f.reduce(lambda a,b: a+b,pos_num,0))
    
print(add(2,3,4,5,-1,2,2))
```

```python
"""Write a code for Hotel Management System using OOPS
1. Create a Room class that has attributes such as room number, room type, rate, and availability (private)
2. Implement methods to book a room, check in a guest, and check out a guest
3. Use encapsulation to hide the room's unique identification number
4. Inherit from the Room class to create a SuiteRoom class and a StandardRoom class, each with their own
specific attributes and methods."""

class Room:
    def __init__(self,room_no,room_type,rate):
        self.room_no=room_no
        self.room_type=room_type
        self.rate=rate
        self.__available=True
        self.__id=None
        
    def get_id(self):
        return self.__id

    def set_id(self, id):
        self.__id = id
    
    def available(self):
        if self.__available:
            return self.__available
        
    def book_room(self):
        if self.available():
            print(f"room {self.room_no} ({self.room_type}) succesfully booked, rupees {self.rate} has been transferred")
            self.__available=False
            
    def check_in(self):
        if not self.available():
            return(f"guest has checked in to room {self.room_no} ({self.room_type})")
        else:
            return("room isn't booked!")
        
    def check_out(self):
        if not self.available():
            print(f"check out succesful! room {self.room_no} ({self.room_type}) is not empty")
            self.__available=True
        else:
            return("room isn't booked!")
        
    def display_info(self):
        print(f"Room ID: {self.get_id()}")
        print(f"Room Number: {self.room_no}")
        print(f"Room Type: {self.room_type}")
        print(f"Rate: {self.rate} per night")
        
        
class SuiteRoom(Room):
    def __init__(self, room_no, rate, occupancy):
        super().__init__(room_no, "Suite", rate)
        self.occupancy = occupancy  # Maximum number of guests

    def display_info(self):
        print("-" * 20)
        super().display_info()
        print(f"Max Occupancy: {self.occupancy} guests")
        print("-" * 20)

class StandardRoom(Room):
    def __init__(self, room_no, rate, bed_size):
        super().__init__(room_no, "Standard", rate)
        self.bed_size = bed_size  # Size of the bed (e.g., King, Queen)

    def display_info(self):
        print("-" * 20)
        super().display_info()
        print(f"Bed Size: {self.bed_size}")
        print("-" * 20)

guest1=SuiteRoom(1,5000,2)
guest1.set_id(1000)
guest1.get_id()
(guest1.book_room())
(guest1.display_info())
(guest1.check_in())
(guest1.check_out())

guest2=StandardRoom(2,2000,"king size")
guest2.set_id(2000)
guest2.get_id()
(guest2.book_room())
(guest2.display_info())
(guest2.check_in())
(guest2.check_out())

    
```

```python
"""8.Write a code for Fitness Club Management System using OOPS
1. Create a Member class that has attributes such as name, age, membership type, and membership status(private)
2. Implement methods to register a new member, renew a membership, and cancel a membership
3. Use encapsulation to hide the member's unique identification number
4. Inherit from the Member class to create a FamilyMember class and an IndividualMember class, each with
their own specific attributes and methods.
"""

class Member:
    def __init__(self, name, age, membership_type):
        self.__id = None  # Encapsulated unique ID
        self.name = name
        self.age = age
        self.membership_type = membership_type
        self.__membership_status = True  # Private attribute for membership status

    def get_id(self):
        return self.__id

    def set_id(self, id):
        self.__id = id

    def is_active(self):
        return self.__membership_status

    def register(self):
        print(f"Member {self.name} successfully registered!")

    def renew_membership(self):
        if self.is_active():
            print(f"Member {self.name}'s membership successfully renewed!")
        else:
            print(f"Member {self.name}'s membership needs to be activated before renewal.")

    def cancel_membership(self):
        if self.is_active():
            print(f"Member {self.name}'s membership cancelled.")
            self.__membership_status = False
        else:
            print(f"Member {self.name}'s membership is already cancelled.")
            
    def display_info(self):
        print(f"Member ID: {self.get_id()}")
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")
        print(f"Membership Type: {self.membership_type}")
        

class FamilyMember(Member):
    def __init__(self, name, age, membership_type, num_family_members):
        super().__init__(name, age, membership_type)
        self.num_family_members = num_family_members

    def display_info(self):
        super().display_info()
        print(f"Number of Family Members: {self.num_family_members}")
        print("-" * 20)

class IndividualMember(Member):
    def __init__(self, name, age, membership_type, trainer):
        super().__init__(name, age, membership_type)
        self.trainer = trainer  # Optional attribute for assigned trainer
        
    def display_info(self):
        super().display_info()
        print(f"Assigned Trainer: {self.trainer if self.trainer else 'None'}")
        print("-" * 20)

member1 = FamilyMember("Dey Family", None, "Family", 3)
member1.set_id(1) 
member1.register()
member1.display_info()

member2 = IndividualMember("Neelanjana", 30, "Individual", "Chloe Ting")
member2.set_id(2)  
member2.register()
member2.display_info()
```

```python
"""9.Write a code for Event Management System using OOPS
1. Create an Event class that has attributes such as name, date, time, location, and list of attendees (private)
2. Implement methods to create a new event, add or remove attendees, and get the total number of attendees
3. Use encapsulation to hide the event's unique identification number
4. Inherit from the Event class to create a PrivateEvent class and a PublicEvent class, each with their own
specific attributes and methods."""

class Event:
    def __init__(self, name, date, time, location):
        self.__id = None 
        self.name = name
        self.date = date
        self.time = time
        self.location = location
        self.__attendees = []
        
    def get_id(self):
        return self.__id
    
    def set_id(self, id):
        self.__id = id

    def add_attendee(self, attendee):
        if attendee not in self.__attendees:
            self.__attendees.append(attendee)
            print(f"{attendee} successfully registered for {self.name}!")
        else:
            print(f"{attendee} is already registered for {self.name}.")

    def remove_attendee(self, attendee):
        if attendee in self.__attendees:
            self.__attendees.remove(attendee)
            print(f"{attendee} successfully unregistered from {self.name}.")
        else:
            print(f"{attendee} is not registered for {self.name}.")

    def get_num_attendees(self):
        return len(self.__attendees)

    def display_info(self):
        print(f"Event ID: {self.get_id()}")
        print(f"Event Name: {self.name}")
        print(f"Date: {self.date}")
        print(f"Time: {self.time}")
        print(f"Location: {self.location}")
        print(f"Number of Attendees: {self.get_num_attendees()}")

class PrivateEvent(Event):
    def __init__(self, name, date, time, location, organizer):
        super().__init__(name, date, time, location)
        self.organizer = organizer

    def display_info(self):
        super().display_info()
        print(f"Organizer: {self.organizer}")
        print("-" * 20)

class PublicEvent(Event):
    def __init__(self, name, date, time, location,chief_guest):
        super().__init__(name, date, time, location)
        self.chief_guest=chief_guest
        
    def display_info(self):
        super().display_info()
        print(f"Chief guest: {self.chief_guest}")
        print("-" * 20)

event1 = PrivateEvent("Community Fair", "2024-08-15", "10:00 AM", "Shanghai","Oogway")
event1.set_id(1)  
event1.add_attendee("Po")
event1.add_attendee("Shifu")
event1.remove_attendee("Tailung")
event1.display_info()

event2 = PublicEvent("Science conference", "2024-09-20", "12:00 PM", "DFC", "Doofensmitz")
event2.set_id(2)  
event2.add_attendee("Bubbles")
event2.add_attendee("Blossom")
event2.add_attendee("Buttercup")
event2.remove_attendee("mojo jojo")
event2.display_info()
```

```python
"""10.Write a code for Airline Reservation System using OOPS
1. Create a Flight class that has attributes such as flight number, departure and arrival airports, departure and
arrival times, and available seats (private)
2. Implement methods to book a seat, cancel a reservation, and get the remaining available seats
3. Use encapsulation to hide the flight's unique identification number
4. Inherit from the Flight class to create a DomesticFlight class and an InternationalFlight class, each with their
own specific attributes and methods."""

class Flight:
    def __init__(self, flight_number, departure_airport, arrival_airport, departure_time, arrival_time, total_seats):
        self.__id = None  # Encapsulated unique ID
        self.flight_number = flight_number
        self.departure_airport = departure_airport
        self.arrival_airport = arrival_airport
        self.departure_time = departure_time
        self.arrival_time = arrival_time
        self.__available_seats = total_seats

    def get_id(self):
        return self.__id

    def set_id(self, id):
        self.__id = id

    def book_seat(self):
        if self.__available_seats > 0:
            self.__available_seats -= 1
            print(f"Seat successfully booked on flight {self.flight_number}!")
        else:
            print(f"Flight {self.flight_number} is fully booked. No seats available.")

    def cancel_reservation(self):
        self.__available_seats += 1
        print(f"Reservation cancelled for flight {self.flight_number}. One seat available again.")

    def get_available_seats(self):
        return self.__available_seats

    def display_info(self):
        print(f"Flight ID: {self.get_id()}")
        print(f"Flight Number: {self.flight_number}")
        print(f"Departure Airport: {self.departure_airport}")
        print(f"Arrival Airport: {self.arrival_airport}")
        print(f"Departure Time: {self.departure_time}")
        print(f"Arrival Time: {self.arrival_time}")
        print(f"Available Seats: {self.get_available_seats()}")
        

class DomesticFlight(Flight):
    def __init__(self, flight_number, departure_airport, arrival_airport, departure_time, arrival_time, total_seats, flight_class):
        super().__init__(flight_number, departure_airport, arrival_airport, departure_time, arrival_time, total_seats)
        self.flight_class = flight_class

    def display_info(self):
        super().display_info()
        print(f"Flight Class: {self.flight_class}")
        print("-" * 20)

class InternationalFlight(Flight):
    def __init__(self, flight_number, departure_airport, arrival_airport, departure_time, arrival_time, total_seats, layover_city=None):
        super().__init__(flight_number, departure_airport, arrival_airport, departure_time, arrival_time, total_seats)
        self.layover_city = layover_city

    def display_info(self):
        super().display_info()
        print(f"Layover City: {(self.layover_city)}")
        print("-" * 20)

flight1 = DomesticFlight("AA123", "LAX", "JFK", "10:00 AM", "02:00 PM", 150, "Economy")
flight1.set_id(1) 
flight1.get_id()
flight1.book_seat()
flight1.cancel_reservation()
flight1.display_info()

flight2=InternationalFlight("BA233", "BST", "LDN", "12:00 AM", "01:00 PM", 350, "DMK")
flight2.set_id(1) 
flight2.get_id()
flight2.book_seat()
flight2.cancel_reservation()
flight2.display_info()
```

```python
"""16. Write a Python program to create a text file named "employees.txt" and write the details of employees,
including their name, age, and salary, into the file."""

def write_employees_to_file(file_path, employees):
    try:
        with open(file_path, 'w') as file:
            for employee in employees:
                name, age, salary = employee
                file.write(f"Name: {name}, Age: {age}, Salary: {salary}\n")
        return "Employee details written successfully."
    except IOError:
        return "Error writing to file."

if __name__ == "__main__":
    employees = [("Mandira", 30, 70000),("Anindita", 24, 48000), ("Mira", 45, 90000),("Rubi", 28, 52000)]

    # Path to the file
    file_path = "employees.txt"

    # Write employee details to the file
    result_message = write_employees_to_file(file_path, employees)
    print(result_message)
```
